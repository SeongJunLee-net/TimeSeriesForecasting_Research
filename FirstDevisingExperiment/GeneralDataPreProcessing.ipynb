```python
import pandas as pd
```

```python
data = pd.read_csv('./GeneralData/ElectricityLoadDiagrams.txt',sep=';')
data
```

### 데이터프레임에서 특정 문자열을 포함하는 행을 찾는 메소드
`DataFrame.str.contains()`

### 데이터프레임에서 특정 문자값을 바꾸는 메소드
`DataFrame.str.replace('찾을문자열','바꿀문자열')`
### 완벽하게 일치하는 문자를 바꾸는 메소드
`DataFrame.replace({'찾을문자열':'바꿀문자열'})`

```python
new_data = data.iloc[:,:11]
new_data
```

```python
for i in range(1,new_data.shape[1]):
    new_data.iloc[:,i] = new_data.iloc[:,i].str.replace(',','.')
    new_data.iloc[:,i] = new_data.iloc[:,i].astype(float)
```

```python
new_data.info()
```

### datetime중에서 특정 시간까지만 추출하기

```python
import datetime as dt
```

```python
new_data['Unnamed: 0'] = pd.to_datetime(new_data['Unnamed: 0'])
```

```python
new_data['Unnamed: 0'] = new_data['Unnamed: 0'].dt.strftime("%Y-%m-%d %H")
```

```python
new_data['Unnamed: 0'] = pd.to_datetime(new_data['Unnamed: 0'])
```

#### DeepAR과 같은 설정으로 2014년 1월1일 이후에 데이터에 대해서만 수행하고 9월1일까지 Training에 이용

```python
new_data = new_data[(new_data['Unnamed: 0'] >= dt.datetime(2014,1,1))]
```

```python
new_data
```

## Groupby+aggregate를 통해서 같은시간에 있는것을 더함

```python
new_data = new_data.groupby(by='Unnamed: 0').sum()
```

```python
new_data
```

## reset_index

```python
new_data.reset_index(drop=True,inplace=True)
```

```python
new_data
```

## rename

```python
for i in range(1,11):
    if i!=10:
        new_data = new_data.rename(columns = {f"MT_00{i}":f"customer{i}"})
    else: new_data = new_data.rename(columns = {f"MT_0{i}":f"customer{i}"})
new_data
```
